```python
nfa_all = pd.read_csv("NFA 2018.csv")
```

```python
#datacleaning
 
one = nfa_all.drop(columns=["ISO alpha-3 code", "UN_subregion", "crop_land", "forest_land", "fishing_ground", "grazing_land", "built_up_land", "total"])
 
three = one.loc[one["record"] == "EFConsPerCap"]
 
four = three.dropna()
```

```python
#creates a function that makes a new table for a specified region
def specified_region(region):
    tbl = four.loc[four['UN_region']== region]
    return tbl
```

```python
#list of region specific tables
asia_region = specified_region('Asia')
africa_region = specified_region('Africa')
northamerica_region = specified_region('North America')
latinamerica_region = specified_region ("Latin America and the Caribbean")
europe_region = specified_region ('Europe')
oceania_region = specified_region('Oceania')
```

```python
#creates a function that selects a column(carbon) from a table for a specified year 

def carbon_year(tbl, year):
    tbl = tbl.loc[tbl['year']== year]
    tbl = tbl[['carbon']]
    #creates series from column
    carbon_series = tbl['carbon'].mean()
    return carbon_series
```

```python
#array of asian mean carbon every four years, then creates a table with column 1 as years and c olumn 2 as carbon
def region_carbon_array(tbl):
    carbon_array = make_array()
    for year in np.arange(1994, 2015, 4):
        variable = carbon_year(tbl, year)
        carbon_array = np.append(carbon_array, variable)
        
    region_carbon_table = Table()
    region_carbon_table = region_carbon_table.with_column('Year', np.arange(1994, 2015, 4))
    region_carbon_table = region_carbon_table.with_column('Carbon', carbon_array)
    region_carbon_scatter = region_carbon_table.scatter('Year', 'Carbon',  fit_line = True)
    plt.ylim(0,7)
    
    return region_carbon_table
```

```python
region_carbon_array(asia_region)
plt.title('Asia');
```
